```python
def plot_df(filename, start, stop ):
    df = read_csv(filename)
    df['index'] = df.index
    df.columns = ['accelx', 'accely', 'accelz', 'gyrox', 'gyroy', 'gyroz', 'index']
    fig, axes = plt.subplots(nrows=3, ncols=3)
    df.iloc[start:stop].plot(x='index', y= 'accelx', ax=axes[0,0], title='accelx')
    df.iloc[start:stop].plot(x='index', y= 'accely',ax=axes[0,1], title='accely')
    df.iloc[start:stop].plot(x='index', y= 'accelz',ax=axes[0,2], title='accelz')
    df.iloc[start:stop].plot(x='index', y= 'gyrox', ax=axes[1,0], title='gyrox')
    df.iloc[start:stop].plot(x='index', y= 'gyroy',ax=axes[1,1], title='gyroy')
    df.iloc[start:stop].plot(x='index', y= 'gyroz',ax=axes[1,2], title='gyroz')
    fig.set_figheight(30)
    fig.set_figwidth(30)
    plt.show()
```

```python
## Buddha Clap
plot_df('raw_data/individual_data/jeff_buddhaClap.csv', 0, 250)
```
